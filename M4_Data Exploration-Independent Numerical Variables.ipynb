```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
Raw_housing_data=pd.read_csv("1. Regression - Module - (Housing Prices).csv")
```

```python
#removing missing values of the sale price:
Raw_housing_data.dropna(inplace=True, axis=0, subset=['Sale Price'])
Raw_housing_data.info()
```

```python
numerical_columns=['No of Bathrooms','Flat Area (in Sqft)','Lot Area (in Sqft)','Area of the House from Basement (in Sqft)',
                  'Latitude','Longitude','Living Area after Renovation (in Sqft)']
```

```python
from sklearn.impute import SimpleImputer
imputer=SimpleImputer(missing_values=np.nan, strategy='median')
Raw_housing_data[numerical_columns]=imputer.fit_transform(Raw_housing_data[numerical_columns])
```

```python
Raw_housing_data.info()
```

```python
imputer=SimpleImputer(missing_values=np.nan, strategy='most_frequent')
Raw_housing_data['Zipcode']=imputer.fit_transform(Raw_housing_data['Zipcode'])
```

```python
Raw_housing_data['Zipcode'].shape
```

```python
# as we can see the column is one-dimensional, so we reshape the column using values.reshape func
#syntax of reshape function is : reshape(rows,columns)
column=Raw_housing_data['Zipcode'].values.reshape(-1,1)
#here -1 signifies that the reshape function will adjust the number of rows by itself
column.shape
```

```python
imputer=SimpleImputer(missing_values=np.nan, strategy='most_frequent')
Raw_housing_data['Zipcode']=imputer.fit_transform(column)
```

```python
Raw_housing_data.info()
```

```python
#Variable Transformations
```

```python
Raw_housing_data['Zipcode']=Raw_housing_data['Zipcode'].astype(object)
Raw_housing_data.dtypes
```

```python
Raw_housing_data['No of Times Visited'].unique()
```

```python
mapping={np.nan:'0',
    'Once':'1',
        'Twice':'2',
        'Thrice':'3',
        'Four':'4'}
Raw_housing_data['No of Times Visited']=Raw_housing_data['No of Times Visited'].map(mapping)
```

```python
Raw_housing_data['No of Times Visited'].unique()
```

```python
Raw_housing_data.info()
```

```python
#creating a new column to check if the house was renowated, as it changes the sales price
Raw_housing_data['Ever Renovated']=np.where(Raw_housing_data['Renovated Year']==0,'No','Yes')
#where there is 0, no will get assigned and where the value is not zero then yes will get asssigned
```

```python
Raw_housing_data.head()
```

```python
#lets write the code to get the year of purchase
Raw_housing_data['Purchase Year']=pd.DatetimeIndex(Raw_housing_data['Date House was Sold']).year
```

```python
#year since renovation=year of sale - year of renovation
Raw_housing_data['Year Since Renovation']=np.where(Raw_housing_data['Ever Renovated']=='Yes', abs(Raw_housing_data['Purchase Year']
                                                                                                 - Raw_housing_data['Renovated Year']),0)
```

```python
Raw_housing_data.head()
```

```python
#now we will delete the unwanted columns 
Raw_housing_data.drop(columns=['Purchase Year','Date House was Sold', 'Renovated Year'],inplace=True)
#axis = 1 as we delete the whole column
```

```python
Raw_housing_data.head()
```

```python
#bymistake created 2 columns, deleting one of them:
Raw_housing_data.drop(columns=['Ever Renovate'],inplace=True)
```
